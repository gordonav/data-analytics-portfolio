
# Background


The following analysis will explore professor salaries using information from a [Kaggle dataset](https://www.kaggle.com/ukveteran/professor-salaries). Professors or postsecondary academic instructors typically hold the highest level of education in a specific field and have salaries that can vary widely based on location, type of institution, competitiveness, and much more. The dataset provided will focus on the following features:

*   Rank(Associate, Assistant, Professor)
*   Discipline(A , B)
*   Years since PhD
*   Years of Service
*   Sex(Male , Female)


# Data

```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#file handling: Salaries.csv
salaries = pd.read_csv('Salaries.csv')
```

```python
#create and set color palette
colors = ['#72aeac', '#729b6f', '#e0c48d', '#729b6f', '#d88e98']
sns.set_palette(sns.color_palette(colors)) 
sns.set_style(style='white')
```

# Analysis

```python
salaries.head(10)
```

```python
headcount = len(salaries['sex'])
print("Headcount: ", headcount) #display headcount
#declare and initialize variables
Males = 0 
Females = 0
for i in salaries['sex']:
  if i == 'Male': 
    Males += 1
  if i == 'Female': 
    Females += 1
#display percentage of male and female employees
print("Males = {} \t {:6.2f}% ".format(Males, (Males/headcount) * 100))
print("Females = {} \t {:6.2f}% ".format(Females, (Females/headcount) * 100))

sns.countplot(x=salaries['sex'])
```

```python
salaries['discipline'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
sns.countplot(x=salaries['sex'],hue=salaries['discipline']).set(title='Sex vs. Discipline');
```

üèÜAssistant Professor: A beginner-level professor that is usually held in a probationary period of five to seven years before being promoted and possibly tenured to associate professor rank.

üèÜAssociate Professor: A mid-level professor that continues to expand research and academic opportunities along with their professor responsibilities. This position is usually granted after the successful publications of research articles, academic papers, or books.

üèÜProfessor: A senior-level professor that continues their responsibilities as a professor but also heavily focuses on research and leadership within departments. This is the highest rank of academic professors.

üìä**In the following analyses, we will explore the 'rank' feature against others in the dataset.**

```python
sns.countplot(x=salaries['rank'],hue=salaries['discipline']).set(title='Sex vs. Rank based on Discipline');
```

```python
sns.countplot(x=salaries['sex'],hue=salaries['rank']).set(title='Sex vs. Rank');
```

```python
plt.figure(figsize=(10, 7))
sns.violinplot(x=salaries['yrs.since.phd'], y=salaries['rank'], hue=salaries['sex'], split=True)
plt.xlabel("Years since PHD")
plt.ylabel("Rank")
plt.title("Rank based on Years since PHD by Sex")
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.violinplot(x=salaries['yrs.service'], y=salaries['rank'], hue=salaries['sex'], split=True)
plt.xlabel("Years of Service")
plt.ylabel("Rank")
plt.title("Rank based on Years of Service by Sex")
plt.show()
```
